```python
import pandas as pd
import openai
import os
import time

# Set up OpenAI API key from environment variable
openai.api_key = # removed my key from online posting for security purposes

# Function to classify descriptions with rate limit handling
def econ_terror(perp_list):
    results = []
    for perpetrator in perp_list:
        prompt = f"Does this terrorist group have an economic motivation? '{perpetrator}' Respond with '1' for yes and '0' for no."
        
        max_retries = 5
        retry_delay = 1
        attempt = 0
        
        while attempt < max_retries:
            try:
                response = openai.ChatCompletion.create(
                    model="gpt-4",
                    messages=[{"role": "user", "content": prompt}]
                )
                classification = response['choices'][0]['message']['content'].strip()
                results.append(int(classification))
                break

            except openai.error.RateLimitError:
                attempt += 1
                print(f"Rate limit reached. Retrying in {retry_delay} seconds...")
                time.sleep(retry_delay)
                retry_delay *= 2
            
            except Exception as e:
                print(f"Error processing description: {perpetrator}\n{e}")
                results.append(None)
                break

    return results

def religious_terror(perp_list):
    results = []
    for perpetrator in perp_list:
        prompt = f"Does this terrorist group have a religious motivation? '{perpetrator}' Respond with '1' for yes and '0' for no."
        
        max_retries = 5
        retry_delay = 1
        attempt = 0
        
        while attempt < max_retries:
            try:
                response = openai.ChatCompletion.create(
                    model="gpt-4",
                    messages=[{"role": "user", "content": prompt}]
                )
                classification = response['choices'][0]['message']['content'].strip()
                results.append(int(classification))
                break

            except openai.error.RateLimitError:
                attempt += 1
                print(f"Rate limit reached. Retrying in {retry_delay} seconds...")
                time.sleep(retry_delay)
                retry_delay *= 2
            
            except Exception as e:
                print(f"Error processing description: {perpetrator}\n{e}")
                results.append(None)
                break

    return results

def political_terror(perp_list):
    results = []
    for perpetrator in perp_list:
        prompt = f"Does this terrorist group have a political motivation? '{perpetrator}' Respond with '1' for yes and '0' for no."
        
        max_retries = 5
        retry_delay = 1
        attempt = 0
        
        while attempt < max_retries:
            try:
                response = openai.ChatCompletion.create(
                    model="gpt-4",
                    messages=[{"role": "user", "content": prompt}]
                )
                classification = response['choices'][0]['message']['content'].strip()
                results.append(int(classification))
                break

            except openai.error.RateLimitError:
                attempt += 1
                print(f"Rate limit reached. Retrying in {retry_delay} seconds...")
                time.sleep(retry_delay)
                retry_delay *= 2
            
            except Exception as e:
                print(f"Error processing description: {perpetrator}\n{e}")
                results.append(None)
                break

    return results

# Apply the function to classify the descriptions and create new columns
df_test['Econ_Terror'] = econ_terror(df_test['Perpetrator'])
df_test['Religious_Terror'] = religious_terror(df_test['Perpetrator'])
df_test['Political_Terror'] = political_terror(df_test['Perpetrator'])

# Save the new dataset to a CSV file
df_test.to_csv('df_test2.csv', index=False)
```
